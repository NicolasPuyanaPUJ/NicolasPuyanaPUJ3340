#### MENOR DE DOS PARES: Escriba una función que devuelva el menor de dos números dados *si* ambos números son pares, pero devuelva el mayor si uno o ambos números son impares
    lesser_of_two_evens(2,4) --> 2
    lesser_of_two_evens(2,5) --> 5

```python
def lesser_of_two_evens(a,b):
    pass
    #Colocar aqui la respuesta
    if a%2==0 and b%2==0:    #Modulos, se usan para saber si un numero es par o impar, pues el modulo de 2 de un numero par siempre es 0 y si es impar es 1, el and obliga a que se cumplan las 2 condiciones para que se realize el if 
        return min(a,b)       #Min devuelve el valor mínimo, y si los 2 valores son pares entonces devuelve el menor de los 2
    else:
        return max(a,b)     #Max devuelve el valor maximo, y si no se cumple la condicion de que los 2 sean pares es porque hay un impar por lo que se devuelve el mayor de los 2
        
        
```

```python
# Revisión
lesser_of_two_evens(2,4)
```

```python
# Revisión
lesser_of_two_evens(2,5)
```

#### PROBLEMA 33: 

Dada una lista de enteros, retornar True si el arreglo tinene un  3  seguido por un  3 .

    has_33([1, 3, 3]) → True
    has_33([1, 3, 1, 3]) → False
    has_33([3, 1, 3]) → False

```python
def has_33(nums):
    pass
    for i in range(len(nums)-1): #Logica for que se recorra la lista hasta el indice -1 para que no se salga de los limites
        if nums[i]==3 and nums[i+1]==3: #si el numero en el indice i es igual a 3 y el numero en el indice i+1 es igual a 3 osea que hay dos valores con el numero 3 consecutivo, se devuelve true
            return True
    return False  # importante la ubicación del return, si este se llegara a incluir dentro del for no funcionaria porque si hay un caso que si se cumple y uno que no va a devovler true y false por lo que se debe colocar fuera del for

```

```python
# Revisión
has_33([1, 3, 3])
```

```python
# Revisión
has_33([1, 3, 1, 3])
```

```python
# Revisión
has_33([3, 1, 3])
```

#### REPLICADOR: Dada una cadena, devuelve una cadena donde por cada carácter en el original hay tres caracteres REPETIDOS
    paper_doll('Hello') --> 'HHHeeellllllooo'
    paper_doll('Mississippi') --> 'MMMiiissssssiiippppppiii'

```python
def paper_doll(text):
  pass
  resultado=""  #Se crea una variable limpia para que se tenga un valor inicial y el tipo de variable que es resultado 
  for c in text:   # c es cada caracter que se encuentra en el texto por lo que recorre cada caracter de la palabra
    resultado += c*3 # lo multiplica y lo va sumando a la cadena para que esta no se sobreescriba y solo termine mostrando el ultimo caracter 3 veces 
  return resultado
  
    

 
```

```python
# Check
paper_doll('Hello')
```

```python
# Check
paper_doll('Mississippi')
```

#### BLACKJACK: Dados tres enteros entre 1 y 11, si su suma es menor o igual a 21, devuelve su suma. Si su suma excede 21 *y* hay un once, reduce la suma total en 10. Finalmente, si la suma (incluso después del ajuste) excede 21, devuelve 'BUST'
    blackjack(5,6,7) --> 18
    blackjack(9,9,9) --> 'BUST'
    blackjack(9,9,11) --> 19

```python
def blackjack(a,b,c):
    pass
    if(sum([a,b,c])<=21):     #se usa la funcion sum con la lista, tambien podria ser una tupla para verificar que sea menor o igual a 21 
     return sum([a,b,c])       #si cumple se devuelve la suma 
    elif(sum([a,b,c])>21 and 11 in [a,b,c]):   #se usa el elif pues no solo hay una condicion si no dos
      return sum([a,b,c])-10              #verifica con el in si algun valor de la lista es igual a 11 y devuelve la suma menos 10 
    else:
      return 'BUST'                         # Si ya execede su valor y no encuentra algun 11 devuelve el bust que significa que se paso 
    
    


```

```python
# Revisión
blackjack(5,6,7)
```
